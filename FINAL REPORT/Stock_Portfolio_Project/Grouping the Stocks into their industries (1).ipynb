```python
import pandas as pd
import yfinance as yf
from collections import defaultdict

# =========================
# LOAD DATA
# =========================
df = pd.read_csv("adjprice.csv")
df_dict = pd.read_csv("data_dictionary.csv")

# Keep original date column
date_col = "Date"

# All ticker columns
ticker_cols = [col for col in df.columns if col != date_col]

# =========================
# CLEAN TICKER NAMES
# =========================
# Example:
# "AAPL US Equity" -> "AAPL"

cleaned_tickers = {}

for col in ticker_cols:
    ticker = col.split(" ")[0]
    cleaned_tickers[col] = ticker

# =========================
# GET SECTOR INFORMATION
# =========================
sector_map = {}

for original_col, ticker in cleaned_tickers.items():
    try:
        stock = yf.Ticker(ticker)
        info = stock.info

        sector = info.get("sector", "Unknown")
        sector_map[original_col] = sector

        print(f"{ticker}: {sector}")

    except Exception as e:
        print(f"Failed for {ticker}: {e}")
        sector_map[original_col] = "Unknown"

# =========================
# GROUP COLUMNS BY SECTOR
# =========================
sector_columns = defaultdict(list)

for col, sector in sector_map.items():
    sector_columns[sector].append(col)

# =========================
# CREATE CSV FOR EACH SECTOR
# =========================
for sector, cols in sector_columns.items():

    # Include Date column
    sector_df = df[[date_col] + cols]

    # Clean filename
    filename = sector.replace("/", "_").replace(" ", "_")

    # Save CSV
    sector_df.to_csv(f"{filename}.csv", index=False)

    print(f"Saved: {filename}.csv")
```
